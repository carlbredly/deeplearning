```python
# Pour supprimer les warnings TensorFlow
from silence_tensorflow import silence_tensorflow
silence_tensorflow()

import time
import keras_cv
import keras
import matplotlib.pyplot as plt
```

# Creation of of stable Diffusion model

```python
model = keras_cv.models.StableDiffusion(
    img_width=512,
    img_height=512,
    jit_compile=False
)
```

# Aries prompt

```python
prompt = "Aries zodiac sign represented as a powerful fire ram with glowing horns, standing on a volcanic mountain, red and orange flames, cosmic sky, fantasy art, ultra detailed, cinematic lighting"
```

# Image generation

```python
start = time.time()

images = model.text_to_image(
    prompt,
    batch_size=3
)

print("Temps de génération :", time.time() - start, "secondes")
```

# show images

```python
def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")

plot_images(images)
```
